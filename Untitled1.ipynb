```python
import pandas as pd
import matplotlib.pyplot as plt

import pandas as pd
import matplotlib.pyplot as plt

def plot_curves(coord_string, threshold=1):
    # Split the coordinate string into individual coordinates
    coordinates = coord_string.split(';')

    # Extract the first two values from each coordinate and convert them to integers
    x_values = [int(coord.split(',')[0]) for coord in coordinates]
    y_values = [int(coord.split(',')[1]) for coord in coordinates]
    
    # Extract x and y coordinates from each pair
    x_coords = [int(coord.split(',')[0]) for coord in coordinates]
    y_coords = [int(coord.split(',')[1]) for coord in coordinates]

    # Calculate movement patterns
    straight_line_detected = False
    straight_line_thresholds = []

    for i in range(2, len(coordinates)):
        x1, y1 = map(int, coordinates[i - 2].split(',')[:2])  # Use only the first two values
        x2, y2 = map(int, coordinates[i].split(',')[:2])       # Use only the first two values

        # Check if the segment forms a straight line
        straight_line_threshold = abs((y2 - y1) * (x_coords[i - 1] - x1) - (x2 - x1) * (y_coords[i - 1] - y1))

        # Modify the condition for detecting a straight line (you need to define a suitable threshold)
        if straight_line_threshold < threshold:
            straight_line_detected = True
        else:
            straight_line_detected = False

        # Store the threshold value for each pair of lines
        straight_line_thresholds.append(straight_line_threshold)

    # Plot the curves
    plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')

    # Highlight the area between two lines with a common dot
    if straight_line_detected:
        max_threshold = min(straight_line_thresholds)
        plt.fill_between(x_values[:-1], y_values[:-1], y2=y_values[1:], color='red', alpha=0.5)#, label=f'Max Threshold: {max_threshold}')
        
        # Add contrast-colored digits to the center of the fill_between
        center_x = sum(x_values[:-1]) / len(x_values[:-1])
        center_y = sum(y_values[:-1]) / len(y_values[:-1])
        plt.text(center_x, center_y, f'{max_threshold}', ha='center', va='center', color='black', fontsize=10, fontweight='bold', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.2'))

    # Add labels and title
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Trajectory from coordinates')
    
    # Add legend
    plt.legend()
    plt.savefig('bot_line_8.pdf')
    # Show the plot
    plt.show()

df_new = pd.read_excel(r'Координаты движения мыши Unix mini.xlsx')

# Example usage:
coordinate_string = df_new['Координаты с отпечатком времени в unix формате (кол-во миллисекунд с 01.01.1970)'][199]#"932,519,1689877037886;989,505,1689877037896;1046,491,1689877037906"
    

# Call the function to plot curves
plot_curves(coordinate_string)
```
